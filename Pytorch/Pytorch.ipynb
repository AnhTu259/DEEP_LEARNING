```python
# BTVN 2
# Tạo một tensor ban đầu có giá trị là 2
# Định nghĩa hàm số và tính gradient
# y = x^3 + 2x^2 + 5x + 1
# Hãy tính dy/dx tại giá trị của x
# Dùng phương pháp Gradient Descent với Learning rate alpha = 0.1 để cập nhật giá trị x trong 10 vòng lặp
```

```python
import torch

# tensor x = 2 và cho phép tính gradient
x = torch.tensor(2.0, requires_grad=True)

# Learning rate
alpha = 0.1

for i in range(10):
    # Định nghĩa hàm y
    y = x**3 + 2*x**2 + 5*x + 1

    # Tính gradient dy/dx
    y.backward()

    # In kết quả mỗi vòng
    print("Vòng:", i+1, ", x =", x.item(), ", dy/dx =", x.grad.item())

    # Gradient Descent: cập nhật x
    with torch.no_grad():
        x -= alpha * x.grad

    # Reset gradient về 0 để vòng sau tính lại
    x.grad.zero_()
```

```python
#BTVN
#Tạo một tập dữ liệu giả lập với x là số giờ học(ngẫu nhiên từ 1-10)và y là số điểm được tính 
#theo công thức y = 3x + 5 + nose với nose là một giá trị ngẫu nhiên nhỏ
#1. khởi tạo tham số w và b ngẫu nhiên với requires_grad = True
#2. Tính MSE
#3. Tính Gradient
#4. Cập nhật tham số w và b bằng Gradient Descent với Learning rate alpha = 0.1
```

```python
import torch
x = torch.randint(1, 11, (20, 1)).float()  # 20 mẫu
noise = torch.randn(20, 1) # Tạo nhiễu ngẫu nhiên
y = 3 * x + 5 + noise
# Tạo tham số Pytorch tính gradient
w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)
# Mô hình hồi quy tuyến tính: y_pred = w*x + b
y_pred = w * x + b
# Hàm mất mát MSE (Mean Squared Error)
loss = torch.mean((y_pred - y) ** 2)
print(loss.item())
# Learning rate (tốc độ học)
alpha = 0.1
loss.backward()
print(loss)
# Cập nhật tham số
with torch.no_grad():
    w -= alpha * w.grad
    b -= alpha * b.grad
# Reset gradient
w.grad.zero_()
b.grad.zero_()

print(w)
print(b)
```

- 269.18328857421875 là giá trị được in bằng loss.item() dạng số thường(float) rất lớn vì w, b là ngẫu nhiên mới trải qua 1 bước học duy nhất.
- tensor(269.1833, grad_fn=<MeanBackward0>) loss được tính bằng hàm mean.
- tensor([17.5695], requires_grad=True) giá trị của w lớn do learning rate = 0.1
- tensor([2.7494], requires_grad=True) b nhỏ hơn w vì b chỉ là hằng số dịch chuyển, còn w nhân với x nên có ảnh hưởng lớn hơn và gradient lớn hơn

```python
#BTVN 
# Bạn tạo giúp thầy tạo tensor với:
# empyty
# zeros
# ones
# random
# reshape với view và view as
```

```python
import torch

a = torch.empty(2, 3) #empyty

b = torch.zeros(2, 3) #zeros

c = torch.ones(2, 3) #ones

d = torch.rand(2, 3) #random

e = torch.arange(6)
e_view = e.view(2, 3) #reshape với view 

f = torch.rand(2, 3)
g = torch.arange(6)
g_view_as = g.view_as(f) #Reshape với view_as

print(a) #Tensor chưa khởi tạo giá trị
print(b) #Tensor toàn số 0
print(c) #Tensor toàn số 1
print(d) #Tensor ngẫu nhiên
print(e) #Tensor 1 chiều
print(e_view) #Tensor đã reshape bằng view
print(f) #Tensor ngẫu nhiên
print(g) #Tensor 1 chiều
print(g_view_as) #Tensor reshape theo shape của f
```
