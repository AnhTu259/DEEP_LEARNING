```python
#BTVN thú vị
#Cho ma trận = [[99,99,99],]
#               [99,99,99],
#               [99,99,99]]

#Hãy tưởng tượng ô thứ 99 trống
#Giả sử 1 là x
#Giả sử 0 là 0

#Lấy đầu vào từ phía X và phí o luân phiên
#Nếu đầu vào từ phía X là (0,0)

#Ma trận sẽ trở thành
#[[x,99,99],
# [99,99,99],
# [99,99,99]]

#Nếu đầu vào từ phía o cũng là (0,0)
#Thông báo họ nhập lại
#Nếu không thì điền vào ma trận

#Thử thách: Nếu người đầu tiên có 3 phần tử liên tử liên tiếp thì dừng trò chơi.
```

```python
board = np.full((3, 3), 99)
def print_board():
    print(board)
    print()
def check_win(player):
    # kiểm tra hàng
    for i in range(3):
        if np.all(board[i, :] == player):
            return True

    # kiểm tra cột
    for i in range(3):
        if np.all(board[:, i] == player):
            return True

    # kiểm tra chéo
    if np.all(np.diag(board) == player):
        return True

    if np.all(np.diag(np.fliplr(board)) == player):
        return True

    return False
current_player = 1  

while True:
    print_board()

    if current_player == 1:
        print("Lượt của X")
    else:
        print("Lượt của O")

    row = int(input("Nhập dòng (0-2): "))
    col = int(input("Nhập cột (0-2): "))

    # kiểm tra ô trống
    if board[row, col] != 99:
        print("Ô này đã được đánh, nhập lại!")
        continue

    # điền vào ma trận
    board[row, col] = current_player

    # kiểm tra thắng
    if check_win(current_player):
        print_board()
        print("X thắng!" if current_player == 1 else "O thắng!")
        break

    # đổi lượt
    current_player = 0 if current_player == 1 else 1

```
